**IMPORTING DATASET**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('songs_normalize.csv')
```

```python
df.info()
```

**Least 10 Popular Songs**

```python
least_popular=df.sort_values(by='popularity').head(10)
print(least_popular['name'])
```

**Most 10 Popular Songs with popularity > 90**

```python
most_popular=df[df['popularity']>90].sort_values(by='popularity',ascending=False).head(10)
print(most_popular['name'])
```

**Set release_date as Index**

```python
df.set_index('release_date',inplace=True)
df.head()
```

**Artist at 25th Row**

```python
print(df.iloc[24]['artists'])
```

**Converting Duration to Seconds**

```python
df['duration_s']=df['duration_ms']/1000
df['duration_s'].head(5)
```

**Correlation between features**

```python
corr_matrix=df.corr()
sns.heatmap(corr_matrix,annot=True)
plt.show()
```

**Creating a sample data Frame**

```python
sample_df=df.sample(frac=0.2)
sample_df.head()
```

**Regression plot between loudness and energy**

```python
sns.regplot(x='loudness',y='energy',data=df)
plt.show()
```

**Regression plot between Acousticness and Popularity**

```python
sns.regplot(x='acousticness',y='popularity',data=df)
plt.show()
```
