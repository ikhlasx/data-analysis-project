```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('retail_price.csv')
sns.violinplot(data=df, x='total_price',palette='coolwarm' )
plt.title('plot for ')
plt.show()
```

```python
sns.boxplot(data=df, x='total_price',y='product_category_name',palette='deep' )
plt.title('plot for total price by category ')
plt.show()
#
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df['total_price'],bins=10,kde=True,color='red')
plt.title('plot for total price')
plt.show()
```

```python
sns.pairplot(df[['product_category_name','qty','total_price']],hue='total_price',palette='Set2')
plt.show()
```

```python
sns.pairplot(df[['product_category_name','qty','total_price']])
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['month_year'])
df.groupby(df['Date'].dt.year)['total_price'].sum().plot(kind='line', figsize=(10, 6))
plt.title('plot for total price by year')
plt.xlabel('year')
plt.ylabel('total price')
plt.show()
```

```python
category=df.groupby(['product_category_name','qty'])['total_price'].sum().unstack()
category.plot(kind='bar',stacked=True,legend=False,figsize=(10,6))
plt.title('plot for total price by category')
plt.xlabel('category')
plt.ylabel('total price')
plt.show()
```

```python
top_sales=df.groupby('product_id')['total_price'].sum().sort_values(ascending=False).head(10)
top_sales.plot(kind='bar',color='red')
plt.title('plot for top sales')
plt.xlabel('product_id')
plt.ylabel('total price')
plt.show()
```

```python
sns.scatterplot(data=df,x='month',y='qty',color='red',)
plt.title('Quantity by Month')
plt.show()
```

```python
category=df.groupby('product_category_name')['total_price'].sum()
category.plot(kind='pie', wedgeprops={'width': 0.3},autopct='%1.1f%%',startangle=90)
plt.title('Total Price by (pie)')
plt.ylabel('')
plt.show()
```

```python
import plotly.express as px
fig = px.treemap(df, path=['product_category_name'], values='total_price',color='product_category_name',color_continuous_scale='viridis')
fig.show()
```

```python
sns.kdeplot(df['total_price'],shade=True,color='red')
plt.title('plot for total price')
plt.xlabel('total price')
plt.ylabel('density')
plt.show()
```

```python
sns.scatterplot(data=df,x='month',y='qty',size='total_price',hue='product_category_name',sizes=(20,200),palette='viridis')
plt.title('Quantity by Month')
plt.show()
```

```python
sns.pairplot(df[['product_category_name','total_price','month']])
plt.title("plot for product category name and total price (scatter plot)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
categories = df[df['product_category_name'] == 'bed_bath_table']
bed_bath_table = categories.groupby("product_id")['total_price'].sum().sort_values(ascending=False).head(5)
sns.barplot(x=bed_bath_table.values, y=bed_bath_table.index, palette='coolwarm')
plt.title('plot for bed bath table')
plt.xlabel('total price')
plt.ylabel('product id')
plt.show()
```
